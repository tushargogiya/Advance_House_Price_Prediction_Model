Now we try to handle missing values
1. In categorical feature
2. IN numerical feature
3. In temporal feature

```python
categorical_with_nan=[feature for feature in categorical_feature if df[feature].isnull().sum()>1]
for feature in categorical_with_nan:
  print('In feature {} we have {} NAN values'.format(feature,df[feature].isnull().mean()))
  df[feature]=df[feature].fillna('missing')
  print('In feature {} we have {} NAN values'.format(feature,df[feature].isnull().mean()))
```

```python
numerical_with_nana=[feature for feature in numerical_feature if df[feature].isnull().sum()>1]
for feature in numerical_with_nana:
  print('In feature {} we have {} NAN values'.format(feature,df[feature].isnull().mean()))
  df[feature+'nan']=np.where(df[feature].isnull(),1,0)
  df[feature]=df[feature].fillna(df[feature].median())
  print('In feature {} we have {} NAN values'.format(feature,df[feature].isnull().mean()))
```

```python
temporal
```

AS in analysis of temporal feature we have seen that respective to YrSold our sale is descreasing but w.r.t to other feature we have some relationship so now to handle  value of year we try to replace these value with YearSold-feature
so that we can get value in digit like 19,20(2022-2010)

```python
for feature in temporal:
  if feature !='YrSold':
    df[feature]=df['YrSold']-df[feature]
```

```python
df.head(10)
```

Now we try to remove those category which have rare value

```python
print(categorical_feature)
```

```python
for feature in categorical_feature:
  temp=df.groupby(feature)['SalePrice'].count()/len(df)
  temp_df=temp[temp>0.01].index
  df[feature]=np.where(df[feature].isin(temp_df),df[feature],'rare_var')
```

```python
df.loc[93]
```

Now we do log_distrinution of skwed value

```python
num_feature=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_feature:
  df[feature]=np.log(df[feature])
```

```python
df.head()
```

Now we will change category feature into OHE

```python
for feature in categorical_feature:
  labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
  labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
  df[feature]=df[feature].map(labels_ordered)
```

```python
df.head()
```

Now we will do scaling of numerical features

```python
scaling_feature=[feature for feature in df.columns if feature not in ['Id','SalePrice']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[scaling_feature])
scaler.transform(df[scaling_feature])
```

```python
data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df[scaling_feature]), columns=scaling_feature)],
                    axis=1)
```

```python
x_train=data.to_csv('X_train',index=False)
```
